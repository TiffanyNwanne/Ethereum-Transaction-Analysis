```python
import pandas as pd
import sqlite3
```

```python
# Load the CSV file into a DataFrame
file_path = '//content/Eth_Txs.csv'
df = pd.read_csv(file_path)
```

```python
# Connect to SQLite3 (in-memory database)
conn = sqlite3.connect(':memory:')
```

```python
# Load DataFrame into SQLite3
df.to_sql('Eth_Txs', conn, index=False, if_exists='replace')
```

```python
# Preview the DataFrame
df.head()
```

```python
# Total Number of Transactions
query = "SELECT COUNT(*) AS total_transactions FROM Eth_Txs;"
pd.read_sql_query(query, conn)
```

```python
# Total Transaction Value
query = "SELECT SUM(value) AS total_transaction_value FROM Eth_Txs;"
pd.read_sql_query(query, conn)
```

```python
# Average Transaction Value
query = "SELECT AVG(value) AS average_transaction_value FROM Eth_Txs;"
pd.read_sql_query(query, conn)
```
